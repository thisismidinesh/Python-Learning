```python
#Experiment with reassigning variables to different types
print("Variable Reassignment Experiment")

# Starting with integer
x = 10
print(f"Initial value of x (integer): {x} - Type: {type(x)}")

# Reassigning to a string
x = "Now I'm a string!"

print(f"Reassigned value of x (string): {x} - Type: {type(x)}")

# Reassigning to a list
x = [1, 2, 3, 4, 5]     

print(f"Reassigned value of x (list): {x} - Type: {type(x)}")

# Reassigning to a boolean
x = True

print(f"Reassigned value of x (boolean): {x} - Type: {type(x)}")

x = 3.14
print(f"Reassigned value of x (float): {x} - Type: {type(x)}")

```

```python
# Task 4: using invalid variable names and observe the errors
print("Task 4: Invalid Variable Names Experiment")

# This section demonstrates invalid variable names
# Uncomment the lines below to see the errors (they are commented out to avoid syntax errors)

print("Invalid variable names (commented out to avoid errors):")
print("# 123variable = 'invalid'  # Cannot start with number")
print("# my-variable = 'invalid'  # Cannot contain hyphens")
print("# class = 'invalid'        # Cannot use reserved keywords")
print("# my variable = 'invalid'  # Cannot contain spaces")
print()

# Task 5: Create variables using different naming conventions
print("Task 5: Different Naming Conventions")

# snake_case (recommended for Python)
user_name = "John Doe"
user_age = 30
is_active_user = True

# camelCase (common in other languages, but not preferred in Python)
userName = "Jane Smith"
userAge = 25
isActiveUser = False

# PascalCase (used for class names)
UserName = "Bob Johnson"
UserAge = 35

# Print all variables
print("snake_case variables:")
print(f"user_name: {user_name}")
print(f"user_age: {user_age}")
print(f"is_active_user: {is_active_user}")
print()

print("camelCase variables:")
print(f"userName: {userName}")
print(f"userAge: {userAge}")
print(f"isActiveUser: {isActiveUser}")
print()

print("PascalCase variables:")
print(f"UserName: {UserName}")
print(f"UserAge: {UserAge}")
print()

# Demonstrate that all naming conventions work
print("All naming conventions are valid in Python, but snake_case is preferred!")
print()

print("=== Variable Implementation Complete ===")
```
