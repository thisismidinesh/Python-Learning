# Built-in Functions in Python

## Introduction

Built-in functions are functions that are always available in Python without needing to import any modules. They provide essential functionality for common programming tasks like input/output, type conversion, mathematical operations, and data manipulation. Python has over 70 built-in functions that are ready to use.

## Key Concepts

### Categories of Built-in Functions
- **Input/Output**: `print()`, `input()`
- **Type Conversion**: `int()`, `float()`, `str()`, `bool()`, `list()`, `dict()`, etc.
- **Mathematical**: `abs()`, `round()`, `min()`, `max()`, `sum()`, `pow()`
- **Sequence Operations**: `len()`, `sorted()`, `reversed()`, `enumerate()`, `zip()`
- **Utility**: `type()`, `isinstance()`, `id()`, `help()`, `dir()`

### Function Characteristics
- Always available (no imports needed)
- Work with various data types
- Return values or perform actions
- Can be combined and nested

## Type Conversion Functions

```python
# Type Conversion Functions Demonstration
print("=== Type Conversion Functions ===")

# int() - convert to integer
print("1. int() conversion:")
print(f"int('42') = {int('42')} (type: {type(int('42'))})")
print(f"int(3.14) = {int(3.14)} (type: {type(int(3.14))})")
print(f"int(True) = {int(True)} (type: {type(int(True))})")

# float() - convert to float
print("\n2. float() conversion:")
print(f"float('3.14') = {float('3.14')} (type: {type(float('3.14'))})")
print(f"float(42) = {float(42)} (type: {type(float(42))})")

# str() - convert to string
print("\n3. str() conversion:")
print(f"str(42) = '{str(42)}' (type: {type(str(42))})")
print(f"str(3.14) = '{str(3.14)}' (type: {type(str(3.14))})")
print(f"str(True) = '{str(True)}' (type: {type(str(True))})")

# bool() - convert to boolean
print("\n4. bool() conversion:")
print(f"bool(0) = {bool(0)}")
print(f"bool(1) = {bool(1)}")
print(f"bool('') = {bool('')}")
print(f"bool('hello') = {bool('hello')}")
print(f"bool([]) = {bool([])}")
print(f"bool([1,2,3]) = {bool([1,2,3])}")

# list(), tuple(), set() - convert to sequences
print("\n5. Sequence conversions:")
my_string = "hello"
my_tuple = (1, 2, 3, 2)
my_list = [1, 2, 3, 2]

print(f"list('{my_string}') = {list(my_string)}")
print(f"tuple({my_list}) = {tuple(my_list)}")
print(f"set({my_tuple}) = {set(my_tuple)}")

# dict() - convert to dictionary
print("\n6. Dictionary conversion:")
pairs = [('a', 1), ('b', 2), ('c', 3)]
print(f"dict({pairs}) = {dict(pairs)}")

# Complex conversions
print("\n7. Complex conversions:")
print(f"list(range(5)) = {list(range(5))}")
print(f"tuple('python') = {tuple('python')}")
print(f"set([1,2,2,3,3,3]) = {set([1,2,2,3,3,3])}")
```

## Mathematical Functions

```python
# Mathematical Functions Demonstration
print("=== Mathematical Functions ===")

# abs() - absolute value
print("1. abs() - absolute value:")
print(f"abs(-5) = {abs(-5)}")
print(f"abs(5) = {abs(5)}")
print(f"abs(-3.14) = {abs(-3.14)}")

# round() - rounding numbers
print("\n2. round() - rounding:")
print(f"round(3.14159) = {round(3.14159)}")
print(f"round(3.14159, 2) = {round(3.14159, 2)}")
print(f"round(3.14159, 3) = {round(3.14159, 3)}")
print(f"round(2.5) = {round(2.5)}")  # Banker's rounding
print(f"round(3.5) = {round(3.5)}")

# min() and max() - minimum and maximum
print("\n3. min() and max():")
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"numbers = {numbers}")
print(f"min(numbers) = {min(numbers)}")
print(f"max(numbers) = {max(numbers)}")

# Multiple arguments
print(f"min(5, 2, 8, 1) = {min(5, 2, 8, 1)}")
print(f"max(5, 2, 8, 1) = {max(5, 2, 8, 1)}")

# With strings
words = ["apple", "Banana", "cherry"]
print(f"words = {words}")
print(f"min(words) = '{min(words)}'")  # lexicographical order
print(f"max(words) = '{max(words)}'")

# sum() - sum of iterables
print("\n4. sum() - summation:")
print(f"sum(numbers) = {sum(numbers)}")
print(f"sum(numbers, start=10) = {sum(numbers, 10)}")  # with start value

# pow() - power function
print("\n5. pow() - power:")
print(f"pow(2, 3) = {pow(2, 3)}")
print(f"pow(2, 0.5) = {pow(2, 0.5)}")  # square root
print(f"pow(10, 2) = {pow(10, 2)}")

# divmod() - division and modulo
print("\n6. divmod() - division and modulo:")
print(f"divmod(17, 3) = {divmod(17, 3)}")  # returns (quotient, remainder)
quotient, remainder = divmod(17, 3)
print(f"17 ÷ 3 = {quotient} with remainder {remainder}")

# Complex mathematical operations
print("\n7. Combined mathematical operations:")
data = [3.14, -2.71, 1.41, -1.73, 2.23]
print(f"Original data: {data}")
print(f"Absolute values: {[abs(x) for x in data]}")
print(f"Rounded to 1 decimal: {[round(x, 1) for x in data]}")
print(f"Min: {min(data)}, Max: {max(data)}, Sum: {sum(data)}")
print(f"Average: {sum(data)/len(data):.2f}")
```

## Sequence Operations

```python
# Sequence Operations Demonstration
print("=== Sequence Operations ===")

# len() - length of sequences
print("1. len() - length:")
my_list = [1, 2, 3, 4, 5]
my_string = "Hello, World!"
my_tuple = (1, 2, 3)
my_dict = {'a': 1, 'b': 2, 'c': 3}

print(f"len({my_list}) = {len(my_list)}")
print(f"len('{my_string}') = {len(my_string)}")
print(f"len({my_tuple}) = {len(my_tuple)}")
print(f"len({my_dict}) = {len(my_dict)}")  # number of key-value pairs

# sorted() - sort sequences
print("\n2. sorted() - sorting:")
unsorted_list = [3, 1, 4, 1, 5, 9, 2, 6]
unsorted_string = "python"

print(f"Original list: {unsorted_list}")
print(f"sorted(list): {sorted(unsorted_list)}")
print(f"sorted(list, reverse=True): {sorted(unsorted_list, reverse=True)}")

print(f"Original string: '{unsorted_string}'")
print(f"sorted(string): {sorted(unsorted_string)}")

# reversed() - reverse sequences
print("\n3. reversed() - reversing:")
print(f"Original list: {my_list}")
print(f"list(reversed({my_list})) = {list(reversed(my_list))}")

# enumerate() - enumerate items with indices
print("\n4. enumerate() - enumeration:")
fruits = ['apple', 'banana', 'cherry']
print(f"Original list: {fruits}")
print("Enumerated:")
for index, fruit in enumerate(fruits):
    print(f"  {index}: {fruit}")

# With custom start
print("Enumerated with start=1:")
for index, fruit in enumerate(fruits, start=1):
    print(f"  {index}: {fruit}")

# zip() - combine sequences
print("\n5. zip() - combining sequences:")
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
cities = ['NYC', 'LA', 'Chicago']

print(f"names: {names}")
print(f"ages: {ages}")
print(f"cities: {cities}")

print("Zipped together:")
zipped = list(zip(names, ages, cities))
print(f"list(zip(names, ages, cities)) = {zipped}")

# Unzipping
print("Unzipping:")
n, a, c = zip(*zipped)
print(f"Names: {list(n)}")
print(f"Ages: {list(a)}")
print(f"Cities: {list(c)}")

# range() - create sequences of numbers
print("\n6. range() - number sequences:")
print(f"list(range(5)) = {list(range(5))}")
print(f"list(range(2, 8)) = {list(range(2, 8))}")
print(f"list(range(1, 10, 2)) = {list(range(1, 10, 2))}")

# all() and any() - check conditions
print("\n7. all() and any() - condition checking:")
numbers = [1, 2, 3, 4, 5]
booleans = [True, True, False, True]

print(f"numbers = {numbers}")
print(f"all(numbers) = {all(numbers)}")  # all non-zero?
print(f"any(numbers) = {any(numbers)}")  # any non-zero?

print(f"booleans = {booleans}")
print(f"all(booleans) = {all(booleans)}")  # all True?
print(f"any(booleans) = {any(booleans)}")  # any True?
```

## Utility Functions

```python
# Utility Functions Demonstration
print("=== Utility Functions ===")

# type() - get object type
print("1. type() - object types:")
print(f"type(42) = {type(42)}")
print(f"type(3.14) = {type(3.14)}")
print(f"type('hello') = {type('hello')}")
print(f"type([1,2,3]) = {type([1,2,3])}")
print(f"type(True) = {type(True)}")

# isinstance() - check object type
print("\n2. isinstance() - type checking:")
x = 42
print(f"x = {x}")
print(f"isinstance(x, int) = {isinstance(x, int)}")
print(f"isinstance(x, float) = {isinstance(x, float)}")
print(f"isinstance(x, (int, float)) = {isinstance(x, (int, float))}")  # multiple types

# id() - get object identity
print("\n3. id() - object identity:")
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"a = {a}")
print(f"b = {b}")
print(f"c = a")
print(f"id(a) = {id(a)}")
print(f"id(b) = {id(b)}")
print(f"id(c) = {id(c)}")
print(f"a is b: {a is b}")  # same object?
print(f"a is c: {a is c}")  # same object?

# dir() - get object attributes
print("\n4. dir() - object attributes:")
my_list = [1, 2, 3]
print(f"dir(list) - first 10 attributes: {dir(my_list)[:10]}")

# help() - get documentation
print("\n5. help() - documentation:")
# help(len)  # Uncomment to see help (commented to avoid long output)

# Basic help info
print("len() function help summary:")
print("- Returns the length (number of items) of an object")
print("- Works with sequences (lists, tuples, strings) and collections (dicts, sets)")

# eval() and exec() - execute code
print("\n6. eval() and exec() - code execution:")
# eval() - evaluate expression
result = eval("2 + 3 * 4")
print(f"eval('2 + 3 * 4') = {result}")

# exec() - execute statement
print("exec() example:")
code = """
x = 10
y = 20
print(f"x + y = {x + y}")
"""
exec(code)

# globals() and locals() - namespace inspection
print("\n7. globals() and locals() - namespaces:")
print(f"Number of global variables: {len(globals())}")
print(f"Some global variables: {list(globals().keys())[:5]}")

# callable() - check if object is callable
print("\n8. callable() - check if callable:")
print(f"callable(len) = {callable(len)}")
print(f"callable(42) = {callable(42)}")
print(f"callable([1,2,3]) = {callable([1,2,3])}")

# hash() - get hash value
print("\n9. hash() - hash values:")
print(f"hash('hello') = {hash('hello')}")
print(f"hash((1,2,3)) = {hash((1,2,3))}")
# print(f"hash([1,2,3]) = {hash([1,2,3])}")  # Lists are not hashable

# ord() and chr() - character encoding
print("\n10. ord() and chr() - character encoding:")
print(f"ord('A') = {ord('A')}")
print(f"ord('a') = {ord('a')}")
print(f"chr(65) = '{chr(65)}'")
print(f"chr(97) = '{chr(97)}'")
```

```python
# Built-in Functions in Python - Implementation

print("=== Built-in Functions Implementation ===")
print()

# Task 1: Input/Output Operations
print("Task 1: Input/Output Operations")
print("-" * 30)

# Simulating user input (in real usage, these would use input())
user_name = "Alice Johnson"  # input("Enter your name: ")
user_age = 28  # int(input("Enter your age: "))
favorite_number = 7  # int(input("Enter your favorite number: "))

print("User Information:")
print(f"Name: {user_name}")
print(f"Age: {user_age}")
print(f"Favorite Number: {favorite_number}")
print(f"Hello {user_name}! You are {user_age} years old and your favorite number is {favorite_number}.")
print()

# Task 2: Type Conversions
print("Task 2: Type Conversions")
print("-" * 25)

# String conversions
str_num = "42"
int_from_str = int(str_num)
float_from_str = float(str_num)

print(f"Original string: '{str_num}'")
print(f"int('{str_num}') = {int_from_str} (type: {type(int_from_str).__name__})")
print(f"float('{str_num}') = {float_from_str} (type: {type(float_from_str).__name__})")

# Number conversions
num = 3.14
int_from_float = int(num)
str_from_float = str(num)

print(f"\nOriginal float: {num}")
print(f"int({num}) = {int_from_float} (type: {type(int_from_float).__name__})")
print(f"str({num}) = '{str_from_float}' (type: {type(str_from_float).__name__})")

# Boolean conversions
print(f"\nBoolean conversions:")
print(f"bool(0) = {bool(0)}")
print(f"bool(1) = {bool(1)}")
print(f"bool('') = {bool('')}")
print(f"bool('hello') = {bool('hello')}")
print()

# Task 3: Mathematical Calculations
print("Task 3: Mathematical Calculations")
print("-" * 35)

numbers = [12, 45, 78, 23, 56, 89, 34, 67, 90, 15]

print(f"Dataset: {numbers}")
print(f"Count: {len(numbers)}")
print(f"Sum: {sum(numbers)}")
print(f"Minimum: {min(numbers)}")
print(f"Maximum: {max(numbers)}")
print(f"Average: {sum(numbers) / len(numbers):.2f}")
print(f"Rounded Average: {round(sum(numbers) / len(numbers), 2)}")

# More complex calculations
squared_numbers = [x**2 for x in numbers]
print(f"Squared numbers: {squared_numbers[:5]}...")  # Show first 5
print(f"Sum of squares: {sum(squared_numbers)}")

# Absolute values and rounding
float_numbers = [3.14159, -2.71828, 1.41421, -1.73205]
print(f"\nFloat numbers: {float_numbers}")
print(f"Absolute values: {[abs(x) for x in float_numbers]}")
print(f"Rounded to 2 decimals: {[round(x, 2) for x in float_numbers]}")
print()

# Task 4: Sequence Manipulations
print("Task 4: Sequence Manipulations")
print("-" * 32)

# List operations
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(f"Original fruits: {fruits}")
print(f"Length: {len(fruits)}")
print(f"Sorted: {sorted(fruits)}")
print(f"Reversed: {list(reversed(fruits))}")

# String operations
text = "Hello, World!"
print(f"\nOriginal text: '{text}'")
print(f"Length: {len(text)}")
print(f"Uppercase: '{text.upper()}'")
print(f"Split: {text.split(', ')}")

# Tuple operations
coordinates = (10, 20, 30, 40, 50)
print(f"\nCoordinates: {coordinates}")
print(f"Length: {len(coordinates)}")
print(f"Maximum: {max(coordinates)}")
print(f"Sum: {sum(coordinates)}")

# Dictionary operations
student_grades = {"Alice": 95, "Bob": 87, "Charlie": 92, "Diana": 88}
print(f"\nStudent grades: {student_grades}")
print(f"Number of students: {len(student_grades)}")
print(f"Highest grade: {max(student_grades.values())}")
print(f"Average grade: {sum(student_grades.values()) / len(student_grades):.2f}")

# Using enumerate
print(f"\nEnumerated fruits:")
for index, fruit in enumerate(fruits, start=1):
    print(f"  {index}. {fruit}")

# Using zip
names = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]
print(f"\nZipped names and scores:")
for name, score in zip(names, scores):
    print(f"  {name}: {score}")
print()

# Task 5: Data Analysis
print("Task 5: Data Analysis")
print("-" * 20)

# Sample dataset
data = [23, 45, 67, 12, 89, 34, 56, 78, 90, 45, 67, 23, 45, 78, 34]

print(f"Dataset: {data}")
print(f"Count: {len(data)}")
print(f"Sum: {sum(data)}")
print(f"Mean: {sum(data) / len(data):.2f}")
print(f"Median: {sorted(data)[len(data)//2]}")
print(f"Min: {min(data)}")
print(f"Max: {max(data)}")
print(f"Range: {max(data) - min(data)}")

# Frequency analysis
unique_values = list(set(data))
print(f"Unique values: {sorted(unique_values)}")
print("Value frequencies:")
for value in sorted(unique_values):
    count = data.count(value)
    print(f"  {value}: {count} times")

# Statistical measures
mean = sum(data) / len(data)
variance = sum((x - mean) ** 2 for x in data) / len(data)
std_dev = variance ** 0.5

print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print()

# Task 6: Utility Functions
print("Task 6: Utility Functions")
print("-" * 24)

# Type checking
values = [42, 3.14, "hello", [1, 2, 3], True, (1, 2), {"a": 1}]

print("Type checking with isinstance():")
for value in values:
    print(f"  {value!r} is {type(value).__name__}: {isinstance(value, type(value))}")

# Object identity
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"\nObject identity:")
print(f"  a = {a}")
print(f"  b = {b}")
print(f"  c = a")
print(f"  a is b: {a is b} (different objects)")
print(f"  a is c: {a is c} (same object)")

# Callable check
objects = [len, 42, "hello", [1, 2, 3], print]

print(f"\nCallable check:")
for obj in objects:
    print(f"  {obj!r} is callable: {callable(obj)}")

# Character encoding
print(f"\nCharacter encoding:")
letters = ['A', 'a', 'Z', 'z', '0', '9']
for letter in letters:
    code = ord(letter)
    back = chr(code)
    print(f"  '{letter}' -> {code} -> '{back}'")

# Hash values
print(f"\nHash values:")
hashable_items = ["hello", (1, 2, 3), 42, True]
for item in hashable_items:
    print(f"  hash({item!r}) = {hash(item)}")

print("\n=== Built-in Functions Implementation Complete ===")
```
