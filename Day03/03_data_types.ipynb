# Python Built-in Data Types (with Details + Examples)

Python has a set of **built-in data types** that determine what kind of values can be stored and what operations can be performed on them.

This notebook includes:
- A clear explanation of each core data type
- Examples you can run immediately
- A practical code section that demonstrates behaviors like mutability and type conversion

---

## Key Data Type Categories

### Numeric Types
- **int**: Integers (whole numbers)
- **float**: Floating-point (decimal numbers)
- **complex**: Complex numbers (real + imaginary)

### Text Type
- **str**: String (a sequence of characters)

### Sequence Types
- **list**: Ordered, mutable collection
- **tuple**: Ordered, immutable collection
- **range**: Sequence of numbers commonly used in loops

### Mapping Type
- **dict**: Key → value mapping with fast lookup

### Set Types
- **set**: Unordered collection of unique elements
- **frozenset**: Immutable set

### Boolean Type
- **bool**: `True` or `False`

### Binary Types
- **bytes**: Immutable sequence of bytes
- **bytearray**: Mutable sequence of bytes
- **memoryview**: View of another binary object without copying

---

```python
# Examples: Core Data Types in Python

# --- Numeric Types ---
print("--- Numeric Types ---")

x = 10          # int
y = -5          # int
z = 0b1010      # int (binary literal)
w = 0xA         # int (hex literal)
print("int examples:", x, y, z, w, "type=", type(x))

pi = 3.14159    # float
sci = 1.23e4    # float (scientific notation)
print("float examples:", pi, sci, "type=", type(pi))

c = 3 + 4j      # complex
print("complex example:", c, "real=", c.real, "imag=", c.imag, "type=", type(c))

# Type conversion
print("int(3.9)->", int(3.9))
print("float(2)->", float(2))
print("complex(1,2)->", complex(1, 2))

```

```python
# --- Strings ---
print("\n--- Strings ---")

single = 'Hello'
double = "World"
multiline = """This is
a multi-line string"""

print(single, double)
print(multiline)
print("Length:", len(double))
print("Upper:", double.upper())
print("Contains 'or':", 'or' in double)
```

```python
# --- List, Tuple, Range ---
print("\n--- Sequences (list, tuple, range) ---")

fruits = ["apple", "banana", "cherry"]
print("List (before):", fruits)
fruits.append("orange")
fruits[1] = "blueberry"
print("List (after):", fruits)

coords = (10, 20, 30)
print("Tuple:", coords)
print("Unpacked:", coords[0], coords[1], coords[2])
print(fruits[:0:-1])  # Reverse list

print("Range:", list(range(0, 10, 2)))
```

```python
# --- Dictionary ---
print("\n--- Dictionary ---")

person = {"name": "Alice", "age": 25, "city": "New York"}
print("Dict:", person)
print("Keys:", list(person.keys()))
print("Values:", list(person.values()))
person["email"] = "alice@example.com"
person["age"] = 26
print("Updated:", person)
```

```python
# --- Set ---
print("\n--- Set ---")

numbers = {1, 2, 3, 2, 1, 3}
print("Set (duplicates removed):", numbers)
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference:", set1 - set2)

set1.add(4)
print("Set1 after adding 4:", set1)

print("union:", set1 | set2)
print("Difference:", set1 - set2)
print("Intersection:", set1 & set2)
```

```python

# --- Boolean + Binary ---
print("\n--- Boolean and Binary Types ---")

flag = True
print("Boolean:", flag, "type=", type(flag))
print("True + 1 ->", True + 1)

b = b"hello"
print("Bytes:", b)

ba = bytearray(b"world")
ba[0] = 87  # 'W'
print("Bytearray:", ba)

mv = memoryview(b)
print("Memoryview first byte:", mv[0])
```

```python
# Data Types in Python - Implementation

print("=== Data Types in Python - Implementation ===")
print()

# Task 1: Create variables of each basic data type
print("Task 1: Creating variables of each basic data type")

# Numeric types
my_int = 42
my_float = 3.14159
my_complex = 2 + 3j

# Text type
my_string = "Hello, Python!"

# Boolean type
my_bool = True

# Sequence types
my_list = [1, 2, 3, 4, 5]
my_tuple = ("apple", "banana", "cherry")
my_range = range(0, 10, 2)

# Mapping type
my_dict = {"name": "Alice", "age": 25, "city": "New York"}

# Set types
my_set = {1, 2, 3, 2, 1}  # Duplicates will be removed

print("Variables created successfully!")
print()
```

```python
# Task 2: Print the type of each variable using type()
print("Task 2: Printing types of each variable")

print(f"my_int: {my_int}, type: {type(my_int)}")
print(f"my_float: {my_float}, type: {type(my_float)}")
print(f"my_complex: {my_complex}, type: {type(my_complex)}")            
print(f"my_string: '{my_string}', type: {type(my_string)}")
print(f"my_bool: {my_bool}, type: {type(my_bool)}")
print(f"my_list: {my_list}, type: {type(my_list)}")
print(f"my_tuple: {my_tuple}, type: {type(my_tuple)}")
print(f"my_range: {my_range}, type: {type(my_range)}")
print(f"my_dict: {my_dict}, type: {type(my_dict)}")
print(f"my_set: {my_set}, type: {type(my_set)}")
print()
```

```python
# Task 3: Demonstrate operations specific to each data type
print("Task 3: Demonstrating operations specific to each data type")

# Numeric operations
print("Numeric operations:")
print(f"my_int + 10 = {my_int + 10}")
print(f"my_float * 2 = {my_float * 2}")
print(f"my_complex.conjugate() = {my_complex.conjugate()}")
print()

# String operations
print("String operations:")
print(f"Length of string: {len(my_string)}")
print(f"Uppercase: {my_string.upper()}")
print(f"Replace 'Python' with 'World': {my_string.replace('Python', 'World')}")
print()

# Boolean operations
print("Boolean operations:")
print(f"not my_bool = {not my_bool}")
print(f"my_bool and False = {my_bool and False}")
print(f"my_bool or False = {my_bool or False}")
print()

# List operations
print("List operations:")
my_list_copy = my_list.copy()
print(f"Original list: {my_list}")
my_list_copy.append(6)  
print(f"Modified list: {my_list_copy}")
my_list_copy[0] = 0
print(f"Modified list after changing first element: {my_list_copy}")    
print(f"Modified copy: {my_list_copy}")
print(f"List slice [1:4]: {my_list[1:4]}")
print()

```

```python
# Tuple operations (limited since immutable)
print("Tuple operations:")
print(f"Tuple length: {len(my_tuple)}")
print(f"First element: {my_tuple}")
print(f"Tuple slice [0:2]: {my_tuple[0:2]}")
print(f"Tuple index of 'banana': {my_tuple.index('banana')}")
print(f"Count 'apple': {my_tuple.count('apple')}")
print()
```

```python

# Range operations
print("Range operations:")
print(f"Range as list: {list(my_range)}")
print(f"Range length: {len(my_range)}")
for i in my_range:
    print(f"Range value:{i}")
```

```python
# Dictionary operations
print("Dictionary operations:")
my_dict_copy = my_dict.copy()
print(f"Original dict: {my_dict}")
my_dict_copy["email"] = "midun@test.com"
print(f"Updated dict copy: {my_dict_copy}")
print(f"Keys: {list(my_dict.keys())}")
print(f"Values: {list(my_dict.values())}")
print(f"Get: {my_dict.get("name")}")
```

```python
# Set operations
print("Set operations:")
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(f"my_set: {my_set}")
print(f"set1 | set2 (union): {set1 | set2}")
print(f"set1 & set2 (intersection): {set1 & set2}")
print(f"set1 - set2 (difference): {set1 - set2}")
my_set.add(6)
print(f"After adding 6: {my_set}")
my_set.remove(2)
print(f"After removing 2: {my_set}")
print()
```

```python
# Task 4: Try to modify an immutable data type (tuple) and observe the error
print("Task 4: Attempting to modify an immutable data type (tuple)")

print("Original tuple:", my_tuple)
try:
    my_tuple[0] = "grape"  # This will raise a TypeError
except TypeError as e:
    print(f"Error: {e}")
```

```python
# Task 5: Create nested data structures
print("Task 5: Creating nested data structures")

# List of dictionaries
print("List of dictionaries:")
students = [
    {"name": "Alice", "age": 25, "grades": [85, 90, 88]},
    {"name": "Bob", "age": 22, "grades": [78, 82, 85]},
    {"name": "Charlie", "age": 24, "grades": [92, 88, 90]}
]

for student in students:
    print(f"Name: {student['name']}, Age: {student['age']}, Grades: {student['grades']}")
```

```python
# Dictionary with list values
print("Dictionary with list values:")
courses = {
    "math": ["Alice", "Charlie", "Eve"],
    "science": ["Bob", "David", "Eve"],
    "history": ["Alice", "Bob", "Frank"]
}

for course, students_list in courses.items():
    print(f"{course.title()}: {', '.join(students_list)} ({len(students_list)} students)")

print()
```

```python
# More complex nested structure
print("Complex nested structure (dictionary containing lists and other dictionaries):")
company = {
    "name": "TechCorp",
    "departments": {
        "engineering": {
            "manager": "Alice",
            "employees": ["Bob", "Charlie", "David"],
            "projects": ["AI System", "Web App"]
        },
        "sales": {
            "manager": "Eve",
            "employees": ["Frank", "Grace"],
            "projects": ["Client Outreach", "Market Analysis"]
        }
    },
    "locations": ["New York", "San Francisco", "London"]
}

print(f"Company Name: {company['name']}")
print(f"Engineering Manager: {company['departments']['engineering']['manager']}")
print(f"Sales Manager: {company['departments']['sales']['manager']}")
print(f"Engineering Projects: {company['departments']['engineering']['projects']}")
print(f"Locations: {company['locations']}")

for dept, info in company["departments"].items():
    print(f"Department: {dept.title()}, Manager: {info['manager']}, Employees: {', '.join(info['employees'])}, Projects: {', '.join(info['projects'])}")        
    
```
